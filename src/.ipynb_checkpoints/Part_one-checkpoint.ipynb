```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
from tqdm import tqdm
import cvxpy as cp
```

## Utility functions

```python
def k(x,y):
  # Returns the result of Euclidiean kernel k(x,y) 
  return np.exp(-(x-y)**2) 

def compute_K(X,Y):
    # Computes the matrix K = (k(x,y))
    return np.array([[k(Xi,Yj) for Yj in Y] for Xi in X])

def plot_f(alpha,x_selected,x):
    # Returns the evaluation of the function f defined by alpha at points x
    K = compute_K(x,x_selected)
    return K@alpha
```

```python
with open('../databases/first_database.pkl', 'rb') as f:
  x,y = pkl.load(f)
```

## Parameters of the problem

```python
n = 100
m = 10
a = 5
sigma = 0.5
M = np.random.choice(np.array(range(n)),size=m, replace=False)

x_selected = x[M]
y_selected = y[M]

agents = [M[2*i:2*i+2] for i in range(a)]
```

## Find the true $\alpha^\star$ using CVX 

```python
def objectif(alpha):
    return cp.pnorm(y[:n] - K_nm@alpha,p=2)**2/(2*sigma**2) + 0.5*cp.quad_form(alpha, K_mm)
```

```python
# Construct the problem.
K_mm = compute_K(x_selected,x_selected)
K_nm = compute_K(x[:n],x_selected)
true_alpha = cp.Variable(m)
objective = cp.Minimize(objectif(true_alpha))
prob = cp.Problem(objective)

# The optimal objective value is returned by `prob.solve()`.
result = prob.solve()
# The optimal value for x is stored in `x.value`.
print(true_alpha.value)
true_alpha = true_alpha.value
```

```python
s = 250
x_prime = np.linspace(-1,1,s)
f_approx_true_cvx = plot_f(true_alpha,x_selected,x_prime)
plt.plot(x_selected,y_selected,'o',label="Selected points")
plt.plot(x_prime,f_approx_true_cvx, label= "$f$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Plot of $f$ using CVX to solve the problem")
```

## Find the true $\alpha^\star$ using linear algebra

We have 
$$\alpha^\star = \arg\min_{\alpha \in \mathbb{R}^m} \frac{1}{2}\alpha^T K_{mm}\alpha + \frac{1}{2\sigma^2}\|y - K_{nm}\alpha\|^2_2$$
Let us denote 
$$g \colon \alpha \mapsto \frac{1}{2}\alpha^T K_{mm}\alpha + \frac{1}{2\sigma^2}\|y - K_{nm}\alpha\|^2_2$$
Then, $g$ is differentiable and one has :
$$\nabla g(\alpha) = K_{mm}\alpha - \frac{1}{\sigma^2}K_{nm}^T(y - K_{nm}\alpha)$$
So,
$$\nabla g(\alpha) = 0 \iff \left[K_{mm} + \frac{1}{\sigma^2}K_{nm}^TK_{nm}\right]\alpha = \frac{1}{\sigma^2}K_{nm}^Ty \iff \alpha = \frac{1}{\sigma^2}\left[K_{mm} + \frac{1}{\sigma^2}K_{nm}^TK_{nm}\right]^{-1}K_{nm}^Ty$$
The question is : is $K_{mm} + \frac{1}{\sigma^2}K_{nm}^TK_{nm}$ invertible ? Manifestement oui !

```python
true_alpha_linalg = np.linalg.solve(K_mm + K_nm.T@K_nm/(sigma**2),K_nm.T@y[:n]/(sigma**2))
```

```python
s = 250
x_prime = np.linspace(-1,1,s)
f_approx_true_linalg = plot_f(true_alpha_linalg,x_selected,x_prime)
plt.plot(x_selected,y_selected,'o',label="Selected points")
plt.plot(x_prime,f_approx_true_linalg, label= "$f_{linalg}$")
plt.plot(x_prime,f_approx_true_cvx, label= "$f_{cvx}$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Plot of $f$ using linear algebra to solve the problem \n Error between both = " + str(np.linalg.norm(f_approx_true_linalg-f_approx_true_cvx)))
```

## Study of the $f_a$

The problem we want to solve in this part is 
$$\alpha^\star = \arg\min_{\alpha \in \mathbb{R}^m} \sum_{a=1}^5 \frac{1}{5}\frac{1}{2}\alpha^T K_{mm}\alpha + \frac{1}{2\sigma^2}\sum_{i\in A}\|y_i - K_{(i)m}\alpha\|^2_2$$

Let us then define, for $a \in \{1,...,5\}$,
$$f_a \colon \alpha \mapsto \frac{1}{5}\frac{1}{2}\alpha^T K_{mm}\alpha + \frac{1}{2\sigma^2}\sum_{i\in A}\|y_i - K_{(i)m}\alpha\|^2_2$$
Therefore, the problem is now : $\alpha^\star = \arg\min_{\alpha \in \mathbb{R}^m} \sum_{a=1}^5 f_a(\alpha)$. To correctly choose the parameters for the Decentralized Gradient Descent (DGD) algorithm, we need to study the smoothness of the $f_a$. 

- First of all, the $f_a$ are convex functions as the sum of convex functions.
- We need to find $L_a$ such that $f_a$ is $L_a$-smooth. For this, let us look at the Hessian matrix $\nabla^2 f_a(\alpha)$ of $f_a$ : 
$$\nabla^2 f_a(\alpha) = \frac{1}{5}K_{mm} + \frac{1}{\sigma^2}\sum_{i\in A} K_{(i)m}K_{(i)m}^T$$
We want $L$ to be such that $0 \preccurlyeq \nabla^2 f_a(\alpha) \preccurlyeq L_a I_m$. To find $L_a$ we are going to look at the eigenvalues of $\nabla^2 f_a(\alpha)$. We need to choose $L_a$ such that all eigenvalues of $\nabla^2 f_a(\alpha)$ are smaller that $L_a$.
Then, we choose $L = \max_{a} L_a$. 

```python
L_a = np.zeros(a)
K = compute_K(x[:n],x[:n])
K_mm = K[M,:][:,M]
for j in range(a):
    # We look at the eigenvalues of each \nabla^2 f_a
    nabla_f_a = K_mm/5 
    for i in agents[j]:
        K_im = K[i,:][M]
        nabla_f_a += K_im@K_im.T/(sigma**2)
    
    eigenval, eigenvec = np.linalg.eig(nabla_f_a)
    L_a[j] = np.max(np.abs(eigenval))
L = np.max(L_a)
print("L = "+str(L))
```

## Decentralized Gradient Descent

```python
def nabla_f(alpha,a,sigma,y,K,M,agents):
    K_mm = K[M,:][:,M]
    
    somme = np.zeros(len(alpha))
    for i in agents[a]:
        K_im = K[i,:][M]
        somme += (y[i] - K_im@alpha)*K_im.T
        
    return (1/5)*K_mm@alpha - somme/(sigma**2)
```

```python
def DGD(nb_it,agents,M,W,K,Y,step_size,sigma,alpha_star):
    nb_agents = len(agents) # Number of agents at play
    m = len(M)
    error = np.zeros((nb_it,nb_agents))
    # We start by initializing the alpha at each devices : alpha_O^i = 0
    # as suggested in the paper "On the Convergence of Decentralized Gradient Descent", Yuanet al. (Theorem 1)
    y = np.zeros((nb_agents,m))
    
    for k in tqdm(range(nb_it)):
        # We update y
        new_y = y.copy()
        
         # Mixing
        mixing = W@y
        
        # Update 
        for a in range(nb_agents):
            grad = nabla_f(y[a],a,sigma,Y,K,M,agents)
            new_y[a] = mixing[a] - step_size*grad
    
        y = new_y.copy()
        error[k] = np.linalg.norm(y - alpha_star,axis=1)
    return y, error
```

```python
nb_it = 10000
K = compute_K(x[:n],x[:n])
step_size = 1/L
res,error = DGD(nb_it,agents,M,W,K,y,step_size,sigma,true_alpha)
```

```python
alpha = np.mean(res,axis=0)
```

## Gradient tracking

```python
def gradient_tracking(nb_it,agents,M,W,K,Y,step_size,sigma,alpha_star):
    nb_agents = len(agents) # Number of agents at play
    m = len(M)
    error = np.zeros((nb_it,nb_agents))
    # We start by initializing the alpha at each devices : alpha_O^i = 0
    # as suggested in the paper "On the Convergence of Decentralized Gradient Descent", Yuanet al. (Theorem 1)
    #y = np.random.random(size=(nb_agents,m))
    #g = np.random.random(size=(nb_agents,m))

    y = np.ones((nb_agents,m))
    g = np.ones((nb_agents,m))
    for k in tqdm(range(nb_it)):
        # We update y
        new_y = y.copy()
        # Mixing
        mixing_y = W@y
        # Update 
        new_y = mixing_y - step_size*g
        
        # We update g
        new_g = g.copy()
        # Mixing
        mixing_g = W@g
        #Update
        for a in range(nb_agents):
            grad_k = nabla_f(y[a],a,sigma,Y,K,M,agents)
            grad_k_plus_1 = nabla_f(new_y[a],a,sigma,Y,K,M,agents)
            new_g[a] = mixing_g[a] + (grad_k_plus_1 - grad_k)
            
        y = new_y.copy()
        g = new_g.copy()
        error[k] = np.linalg.norm(y - alpha_star,axis=1)
    return y, error
```

```python
nb_it = 1000
K = compute_K(x[:n],x[:n])
step_size = 1/L
res,error = gradient_tracking(nb_it,agents,M,W,K,y,step_size,sigma,true_alpha)
```

```python
alpha = np.mean(res,axis=0)
```
